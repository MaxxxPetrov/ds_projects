# Описание данных

* ***Технологический процесс***

    * `Rougher feed` — исходное сырье

    * `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant

    * `Xanthate` — ксантогенат (промотер, или активатор флотации)

    * `Sulphate` — сульфат (на данном производстве сульфид натрия);

    * `Depressant` — депрессант (силикат натрия).

    * `Rougher process` (англ. «грубый процесс») — флотация

    * `Rougher tails` — отвальные хвосты

    * `Float banks` — флотационная установка

    * `Cleaner process` — очистка

    * `Rougher Au` — черновой концентрат золота

    * `Final Au` — финальный концентрат золота
    
    

* ***Параметры этапов***

    * `air amount` — объём воздуха
    
    * `fluid levels` — уровень жидкости
    
    * `feed size` — размер гранул сырья
    
    * `feed rate` — скорость подачи
    
    
    
* ***Наименование признаков***

   * Наименование признаков должно быть такое: `[этап].[тип_параметра].[название_параметра]`

   * Пример: `rougher.input.feed_ag`
   
   * Возможные значения для блока `[этап]`:
   
     * `rougher` — флотация
     
     * `primary_cleaner` — первичная очистка
     
     * `secondary_cleaner` — вторичная очистка
     
     * `final` — финальные характеристики
     
  * Возможные значения для блока `[тип_параметра]`:
  
     * `input` — параметры сырья
     
     * `output` — параметры продукта
     
     * `state` — параметры, характеризующие текущее состояние этапа
     
     * `calculation` — расчётные характеристики

Эффективность обогащения руды расчитывается по формуле:



где:
* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.


Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

 Коэффициент восстановления золота находится в столбцах `'rougher.output.recovery'` и `'final.output.recovery'`. С помощью формулы выше расчитаем эффективность и сравним с табличными показателями. Для подсчета ошибки используем MAE.

```python
# стадия обогащения
CR = gold_train['rougher.output.concentrate_au']
FR = gold_train['rougher.input.feed_au']
TR = gold_train['rougher.output.tail_au']

# финальная стадия
CF = gold_train['final.output.concentrate_au']
FF = gold_train['rougher.input.feed_au']
TF = gold_train['final.output.tail_au']
```

```python
rougher_output_recovery_calculate = (CR * (FR - TR)) / (FR * (CR - TR)) * 100

final_output_recovery_calculate = (CF * (FF - TF)) / (FF * (CF - TF)) * 100
```

```python
# gold_train['rougher.output.recovery'].head()
```

```python
# rougher_output_recovery_calculate[:5]
```

```python
# ошибка на rougher стадии
mae_r = mean_absolute_error(gold_train['rougher.output.recovery'],
                          rougher_output_recovery_calculate)

mae_r
```

```python
# ошибка на финальной стадии
mae_f = mean_absolute_error(gold_train['final.output.recovery'],
                          final_output_recovery_calculate)
mae_f
```

MAE между расчетной эффективностью и табличной около нуля, а значит целевые признаки расчитаны правильно.

```python
# переделываем дату в флоат во всех колонках

for gold in [gold_full, gold_train, gold_test]:
    gold['date'] = pd.to_datetime(gold['date'], format='%Y-%m-%d %H:%M:%S') # сначала переводим в формат datetime
    gold['timedelta'] = (gold['date'] - gold_full['date'].min()) / pd.to_timedelta('1 days') # делим значения на интервал 1 день

gold = [gold.drop(columns='date', inplace=True) for gold in [gold_full, gold_train, gold_test]] # убираем колонку date
```

#### Линейная зависимость признаков

Если есть 2 признака которые сильно между собой коррелируют, то лучше такие признаки убрать чтобы избежать мультиколлинеарности. Будем считать, что если коеффициент корреляции больше 0,9 то такие признаки линейно зависимы.

```python
sns.heatmap(features_train.corr()) # тепловая карта корреляции
None
```

```python
features_train.corr() # матрица корреляции
```

Признаков много, глазами все посмотреть довольно трудно, поэтому делегируем эту задачу питону

```python
# булев массив линейно независимых фичей

~(features_train.corr()[(features_train.corr() > 0.9) & (features_train.corr() != 1)].any(axis=1)) 
```

Получили такой массив. Если стоит True, то это значит что данный признак ни с одним из других признаков не имеет линейную зависимость (мы условились что лин зависимость начинается с кэффа 0,9), в противном случае данный признак коррелирует с какими то другими признаками, а значит его нужно убрать.

## Модели

### Smape метрика

Для подсчета качества модели используем метрику sMAPE:

Итоговая метрика складывается из 2 величин:

```python
# функция подсчета smape

def smape(target, predict):
    return (2 * abs(target - predict) / (abs(target) + abs(predict))).sum() * (100 / len(target))
```

### Изучение моделей

```python
# функция проверки моделей

def calc_smape(est, features, target, splits=5):
    
    smapes=[]
    kf = KFold(random_state=12345, shuffle=True, n_splits=splits)
    
    for i, (train_index, test_index) in enumerate(kf.split(features)):
        model = MultiOutputRegressor(est)
        model.fit(features.loc[train_index], target.loc[train_index])
        predicts = model.predict(features.loc[test_index])
    
        smap = (0.25 * smape(target.loc[test_index].iloc[:, 0], predicts[:, 0]) + 
            0.75 * smape(target.loc[test_index].iloc[:, 1], predicts[:, 1]))
    
        smapes.append(smap)

    return sum(smapes) / len(smapes)
```

#### Dummy Model (Baseline)

Данная модель нужна чтобы проверять другие модели на адекватность. Если sMAPE модели будет больше чем у Дамми предиктора, то значит такая модель плохо работает.

```python
print('sMAPE DummyModel:', calc_smape(DummyRegressor(), features_train, target_train)) # baseline
```

#### Linear Regression

```python
print('sMAPE Linear Regression:', calc_smape(LinearRegression(), features_train, target_train))
```

#### Decision Tree

```python
print('sMAPE Decision Tree')
result = 100
dep = 0
for depth in range(5, 16):
    res = calc_smape(DecisionTreeRegressor(max_depth=depth, random_state=1234), features_train, target_train)
    if res < result:
        result = res
        dep = depth
        
    print(f'max_depth = {depth}: {res:.2f}')
    
print()
print('Best result:')
print(f'sMAPE = {result:.2f}, max_depth = {dep}')
```

#### KNeighborsRegressor (ближайшие соседи)

```python
print('sMAPE KNeighbors')
result = 100
n = 0
for neigh in range(2, 11):
    res = calc_smape(KNeighborsRegressor(n_neighbors=neigh), features_train, target_train)
    if res < result:
        result = res
        n = neigh
        
    print(f'n_neighbors = {neigh}: {res:.2f}')
    
print()
print('Best result:')
print(f'sMAPE = {result:.2f}, n_neighbors = {n}')
```
